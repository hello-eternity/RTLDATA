```python
import pandas as pd
import re

def extract_modules(text):
    # 按endmodule切分文本
    segments = text.split('endmodule')
    modules = []

    for segment in segments[:-1]:  # 最后一个段落不含endmodule，所以跳过
        # 在当前段落中找到最后一个module的位置
        module_occurrences = [m.start() for m in re.finditer(r'\bmodule\b', segment)]
        if module_occurrences:
            start = module_occurrences[-1]  # 最后一个module的位置
            # 提取module到endmodule的内容
            module_content = segment[start:] + 'endmodule'
            modules.append(module_content)

    return modules

# 读取CSV文件
df = pd.read_csv('/root/dataset/Verilog_bigquery_GitHub.csv')

# 初始化一个空列表来存储提取的内容
extracted_content = []

# 遍历每行文本，提取并处理
for text in df['text']:
    modules = extract_modules(text)
    extracted_content.extend(modules)

# 将提取的内容转换为DataFrame
extracted_df = pd.DataFrame(extracted_content, columns=['module_content'])

# 将提取的内容写入新的CSV文件
extracted_df.to_csv('extracted_modules.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def estimate_subword_tokens(text):
    # 首先按空格分割
    words = text.split()

    # 初始化token计数
    token_count = 0

    # 遍历每个单词
    for word in words:
        # 假设每个单词至少为一个token
        token_count += 1

        # 可以添加更多的分割逻辑，例如对连字符、撇号等进行分割
        if '-' in word:
            token_count += word.count('-')

    return token_count

# 读取CSV文件
df = pd.read_csv('/root/Verilog_data/extracted_modules.csv')

# 计算每行的token数量
df['token_count'] = df['module_content'].apply(estimate_subword_tokens)

# 定义bins和标签
bins = [0, 2048, 4096, 8192, 16384, float('inf')]
labels = ['<2048', '2048-4096', '4096-8192', '8192-16384', '>16384']

# 将token计数划分到定义的bins
df['token_category'] = pd.cut(df['token_count'], bins=bins, labels=labels, include_lowest=True)

# 统计每个区间的频率
token_distribution = df['token_category'].value_counts().reindex(labels)

# 绘制直方图
plt.figure(figsize=(10, 6))

token_distribution.plot(kind='bar', color='skyblue', edgecolor='black')
# plt.title('Subword-level Token Count Distribution')
plt.xlabel('Token Count', fontsize=30)
plt.ylabel('Frequency', fontsize=30)
plt.xticks(rotation=0,fontsize=14)
plt.yticks(fontsize=14)
plt.savefig("./token.png")
plt.show()
```

```python
import os
import pandas as pd

def read_first_file_in_directory(directory):
    """
    Read the content of the first file in the given directory.
    """
    try:
        for file_name in os.listdir(directory):
            file_path = os.path.join(directory, file_name)
            if os.path.isfile(file_path):
                with open(file_path, 'r') as file:
                    return file.read()
    except Exception as e:
        print(f"Error reading file in {directory}: {e}")
        return None

def process_design_directories(input_root, output_root):
    """
    Process each design directory in the input and output roots.
    """
    data = {'instruction': [], 'input': [], 'output': []}

    for design in os.listdir(input_root):
        input_dir = os.path.join(input_root, design)
        output_dir = os.path.join(output_root, design)

        if os.path.isdir(input_dir) and os.path.isdir(output_dir):
            input_content = read_first_file_in_directory(input_dir)
            output_content = read_first_file_in_directory(output_dir)

            if input_content is not None and output_content is not None:
                data['instruction'].append("Please act as a professional verilog designer.")
                data['input'].append(input_content)
                data['output'].append(output_content)

    return pd.DataFrame(data)

def main():
    # Define the root paths for input and output directories
    input_root = '/root/dataset/testbench/filtered-RTLLM-main/input'
    output_root = '/root/dataset/testbench/filtered-RTLLM-main/output'

    # Process the directories and create a DataFrame
    df = process_design_directories(input_root, output_root)

    # Writing to a CSV file
    output_csv_path = 'verilog_design_tasks.csv'
    df.to_csv(output_csv_path, index=False)
    print(f"CSV file created at: {output_csv_path}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取两个CSV文件
df1 = pd.read_csv('/root/Verilog_data/extracted_modules.csv')
df2 = pd.read_csv('/root/Verilog_bigquery_GitHub.csv')

# 第一个数据集的处理
def estimate_subword_tokens(text):
    words = text.split()
    token_count = 0
    for word in words:
        token_count += 1
        if '-' in word:
            token_count += word.count('-')
    return token_count

df1['token_count'] = df1['module_content'].apply(estimate_subword_tokens)
bins = [0, 2048, 4096, 8192, 16384, float('inf')]
labels = ['<2048', '2048-4096', '4096-8192', '8192-16384', '>16384']
df1['token_category'] = pd.cut(df1['token_count'], bins=bins, labels=labels, include_lowest=True)
token_distribution = df1['token_category'].value_counts().reindex(labels)

# 第二个数据集的处理
counter = {i: 0 for i in range(1, 11)}
for text in df2['text']:
    num_pairs = min(text.count("module"), text.count("endmodule"))
    if 1 <= num_pairs <= 10:
        counter[num_pairs] += 1
pair_counts = list(counter.values())
labels_pairs = list(counter.keys())

# 创建图表
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), dpi=300)

# 定义柱子的宽度和颜色
bar_width = 0.35
color = '#6C8EBF'

# 第一个图表
ax1.bar(token_distribution.index, token_distribution.values, width=bar_width, color=color, edgecolor='black')
ax1.set_xlabel('Token Count', fontsize=20, fontweight='bold')
ax1.set_ylabel('Frequency', fontsize=20, fontweight='bold')
ax1.set_xticks(np.arange(len(labels)))
ax1.set_xticklabels(labels, fontsize=12)
ax1.set_yticklabels(ax1.get_yticks(), fontsize=12)

# 第二个图表
ax2.bar(labels_pairs, pair_counts, width=bar_width, color=color, edgecolor='black')
ax2.set_xlabel('Number of module-endmodule Pairs', fontsize=20, fontweight='bold')
ax2.set_ylabel('Frequency', fontsize=20, fontweight='bold')
ax2.set_xticks(labels_pairs)
ax2.set_xticklabels(labels_pairs, fontsize=12)
ax2.set_yticklabels(ax2.get_yticks(), fontsize=12)

# 调整布局和显示图表
plt.tight_layout()
plt.savefig("./combined_chart.png")
plt.show()
```
